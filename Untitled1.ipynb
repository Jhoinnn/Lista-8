Importação e Visualização Inicial


```python
!pip install --upgrade kneed
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from kneed import KneeLocator


# Carregar a base de dados
base = pd.read_csv('/content/Iris.csv', delimiter=',', encoding='cp1252')
print(base.head())

# Separar as features
Entrada = base.iloc[:, 0:4].values

# Visualização inicial
sns.pairplot(base, hue='class')
plt.show()
```

Normalização dos Dados


```python
# Normalização MinMax
scaler = MinMaxScaler()
Entrada = scaler.fit_transform(Entrada)
print(Entrada)
```

Identificação de Outliers


```python
# Identificação de outliers usando boxplot
plt.figure(figsize=(10, 7))
sns.boxplot(data=base.iloc[:, 0:4])
plt.show()

# Remoção de outliers (se necessário)
Q1 = base.iloc[:, 0:4].quantile(0.25)
Q3 = base.iloc[:, 0:4].quantile(0.75)
IQR = Q3 - Q1

# Filtrar outliers
base_filtered = base[~((base.iloc[:, 0:4] < (Q1 - 1.5 * IQR)) |(base.iloc[:, 0:4] > (Q3 + 1.5 * IQR))).any(axis=1)]
Entrada_filtered = scaler.fit_transform(base_filtered.iloc[:, 0:4].values)
```

Determinação do Número Ideal de Clusters
Método do Cotovelo (Elbow)

O método Elbow (cotovelo) é utilizado para encontrar o número ótimo de clusters em um dataset. Ele envolve plotar a soma das distâncias quadráticas dentro dos clusters (within-cluster sum of squares, WCSS) contra o número de clusters
𝑘
k

Ao plotar o WCSS para diferentes valores de
𝑘
k, geralmente haverá uma diminuição rápida de WCSS com o aumento do número de clusters. No entanto, a taxa de diminuição diminui a partir de certo ponto, formando um "cotovelo" no gráfico. Este ponto de inflexão é considerado o número ótimo de clusters, pois adiciona-se pouco valor adicional ao aumentar mais os clusters.

```python
wcss = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=10)
    kmeans.fit(Entrada)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 7))
plt.plot(range(2, 11), wcss)
plt.xticks(range(2, 11))
plt.title('Método do Cotovelo')
plt.xlabel('Número de Clusters')
plt.ylabel('WCSS')
plt.show()

# Encontrar o ponto de cotovelo
kl = KneeLocator(range(2, 11), wcss, curve='convex', direction='decreasing')
print(f'Número ideal de clusters: {kl.elbow}')
```

Análise da Silhueta

Silhouette Score: Mede o quão semelhante um ponto é ao seu próprio cluster em comparação com outros clusters. A fórmula para a Silhouette de um ponto
𝑖
i é dada por:

s(i)= b(i)−a(i)/max (a(i),b(i))

​


s(i) varia de -1 a 1:

Valores próximos de 1 indicam que o ponto está bem dentro do seu próprio cluster.
Valores próximos de 0 indicam que o ponto está no limite entre dois clusters.
Valores negativos indicam que o ponto pode estar mal atribuído, pois está mais próximo de um cluster diferente.
​


```python
limit = int((Entrada.shape[0] // 2) ** 0.5)
silhouette_scores = []

for k in range(2, limit + 1):
    model = KMeans(n_clusters=k)
    model.fit(Entrada)
    pred = model.predict(Entrada)
    score = silhouette_score(Entrada, pred)
    silhouette_scores.append(score)
    print(f'Silhouette Score k = {k}: {score:.3f}')

plt.figure(figsize=(10, 7))
plt.plot(range(2, limit + 1), silhouette_scores)
plt.xticks(range(2, limit + 1))
plt.title('Análise da Silhueta')
plt.xlabel('Número de Clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

Agrupamento com o Número Ideal de Clusters


```python
# Número ideal de clusters obtido pelo método do cotovelo e silhueta
n_clusters = kl.elbow

kmeans = KMeans(n_clusters=n_clusters, random_state=0)
saida_kmeans = kmeans.fit_predict(Entrada)

# Visualização dos clusters
plt.figure(figsize=(10, 7))
colors = ['purple', 'orange', 'green', 'pink', 'black', 'white']
for i in range(n_clusters):
    plt.scatter(Entrada[saida_kmeans == i, 0], Entrada[saida_kmeans == i, 1], s=100, c=colors[i], label=f'Cluster {i}')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='red', label='Centroides')
plt.legend()
plt.show()
```

O índice Davies-Bouldin é uma medida de avaliação de clusters, onde menores valores indicam clusters mais compactos e bem separados.

Menor Valor: Um menor valor do DBI indica uma melhor separação entre os clusters e maior coesão dentro dos clusters. Idealmente, você busca minimizar o DBI ao escolher o número de clusters.
Maior Valor: Um maior valor do DBI sugere que os clusters são menos bem definidos, indicando sobreposição entre clusters ou alta dispersão dentro dos clusters.

```python
from sklearn.metrics import davies_bouldin_score

dbs = davies_bouldin_score(Entrada, saida_kmeans)
print(f'Índice Davies-Bouldin: {dbs:.3f}')
```

Visualização dos Agrupamentos Incorretos


```python
# Adicionando a classificação real e a predição do K-means na base
base['kmeans'] = saida_kmeans
base['class_id'] = base['class'].astype('category').cat.codes

# Visualizando os agrupamentos incorretos
incorrect = base[base['kmeans'] != base['class_id']]
print(incorrect)

# Visualização gráfica
sns.scatterplot(data=base, x=base.iloc[:, 0], y=base.iloc[:, 1], hue='kmeans', style='class', palette='deep')
plt.title('Agrupamentos do K-means vs Classes Reais')
plt.show()
```
